# Hello World in Zipline backtester
## 1. Dependencies

```python
# zipline's magic command for enabling support to run in jupyter notebook
%load_ext zipline
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

from zipline.api import order_percent, order_target_percent, order, record, symbol, set_benchmark, get_order, \
    schedule_function, date_rules, time_rules, get_open_orders
from zipline.utils.calendars import get_calendar
from zipline.finance.execution import MarketOrder
from zipline import run_algorithm
from datetime import datetime
```

## 2. Define Zipline backtest functions
Below is the simplest example on how to define a strategy and backtest with zipline  

TODO: using the custom dataset ingested in the notebook `0 - Preparing custom dataset for Zipline`

```python
def initialize(context):
    set_benchmark(symbol('AAPL'))

def handle_data(context, data):
    order(symbol('AAPL'), 1)
    record(AAPL=data.current(symbol('AAPL'), 'close'))
    
def analyze(context, perf):
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,8))
    ax1.plot(perf.portfolio_value)
    ax1.set_ylabel('portfolio $ value')
    ax1.legend(loc=0)
    
    ax2.plot(perf.AAPL)
    ax2.set_ylabel('Symbol price')
    plt.show()  
```

## 3. Backtest

```python
# perform the backtest inline in jupyter notebook
#%zipline --bundle fxcm --trading-calendar forex --start 2008-1-3 --end 2018-1-1 -o backtest.pickle --data-frequency daily
%zipline --bundle quantopian-quandl --start 2000-1-1 --end 2018-1-1 -o backtest.pickle
```

```python
# another way of performing a backtest
start_date = pd.to_datetime('2001-01-01').tz_localize('US/Eastern')
end_date = pd.to_datetime('2018-01-01').tz_localize('US/Eastern')
perf = run_algorithm(initialize=initialize, handle_data=handle_data, analyze=analyze, data_frequency='daily', capital_base=10000, bundle='quantopian-quandl', start=start_date, end=end_date)
print(perf.returns)
```

```python
# the backtest result can also be loaded from disk and analyzed
df = pd.read_pickle("backtest.pickle")
df.portfolio_value.plot()
plt.show()
```

## 4. Simple moving average strategy

```python
def initialize(context):
    context.aapl = symbol('AAPL')
    set_benchmark(context.aapl)
    schedule_function(ma_crossover_handling, date_rules.every_day(), time_rules.market_open(hours=1))
    

def analyze(context, perf):
    f, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,8))
    ax1.plot(perf.AAPL, label='price', c='black')
    ax1.plot(perf.ma20, label='ma20', c='b')
    ax1.plot(perf.ma50, label='ma50', c='r')
    ax1.scatter(perf.index[perf.signal == 1], perf.AAPL[perf.signal == 1], label='ma50', c='g', marker='^')
    ax1.scatter(perf.index[perf.signal == -1], perf.AAPL[perf.signal == -1], label='ma50', c='r', marker='v')
    ax2.legend(loc=0)
    ax1.set_ylabel('AAPL chart')
    
    ax2.plot(perf.portfolio_value)
    ax2.set_ylabel('portfolio $ value')
    ax2.legend(loc=0)
    
    plt.show()  
 
  
def handle_data(context,data):
    record(leverage=context.account.leverage)
    record(AAPL=data.current(symbol('AAPL'), 'close'))
    
    hist = data.history(context.aapl,'price', 50, '1d')
    ma50 = hist.mean()
    ma20 = hist[-20:].mean()
    
    record(ma20=ma20)
    record(ma50=ma50)
        
def ma_crossover_handling(context,data):
    hist = data.history(context.aapl,'price', 51, '1d')
    ma50 = hist[-50:].mean()
    ma20 = hist[-20:].mean()
    
    ma50_prev = hist[-51:-1].mean()
    ma20_prev = hist[-21:-1].mean()
    
    open_orders = get_open_orders()
    
    if ma20 > ma50 and ma20_prev < ma50_prev:
        if context.aapl not in open_orders:
            order_target_percent(context.aapl, 1.0)
            record(signal=1)
    elif ma20 < ma50 and ma20_prev > ma50_prev:
        if context.aapl not in open_orders:
            order_target_percent(context.aapl, 0.0)
            record(signal=-1)
    else:
        record(signal=0)
```
